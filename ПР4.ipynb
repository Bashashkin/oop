Построить класс 1-го уровня с указанными в задании полями и методами:
- конструктор,
- функция, которая определяет “качество” (Q ) по заданной формуле,
- вывод информации об объекте.
Построить класс-потомок, который содержит дополнительное поле P и
перекрывает функцию качества (Q ), выполняя вычисление по новой формуле

```python
from datetime import datetime

class Book:
  def __init__(self, name, pages, price):
    self.name = name
    self.pages = pages
    self.price = price
  def q(self):
    return self.price/self.pages

class ProgrammingBook(Book):
  def __init__(self, name, pages, price, publication_year):
    self.name = name
    self.pages = pages
    self.price = price
    self.publication_year = publication_year
  def qp(self):
    a = ProgrammingBook.q(self)
    t = datetime.now().year
    return a-0.2*(t - self.publication_year)
                    

try:
  while True:
    name = input('Введите название книги по программированию')
    if name == "":
      print("Название не должно быть пустым")
      break
    pages = int(input('Введите количество страниц'))
    if pages < 1:
      print('Ошибка! Страниц не может быть меньше 1')
      break
    price = int(input('Введите стоимость книги'))
    if price < 0:
      print('Ошибка! Цена не может быть отрицательной ')
      break
    publication_year = int(input('Введите год издания'))
    if publication_year < 0:
      print('Ошибка! Год издания не может быть таким ранним')
      break
    book2 = ProgrammingBook(name, pages, price, publication_year)
    print("Результат работы функции: ",book2.qp())
    break

except ValueError:
  print('Проверьте данные!')
except ZeroDivisionError:
  print('На ноль делить нельзя!')

```
